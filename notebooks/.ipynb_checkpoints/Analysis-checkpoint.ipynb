```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('input/train_1.csv').fillna(0)
train_df.head()
```

```python
train_df.info()
```

```python
import re
def get_language(url):
    res = re.search('[a-z][a-z].wikipedia.org',url)
    if res:
        return res[0][0:2]
    return 'na'
```

```python
train_df['lang'] = train_df.Page.map(get_language)
```

```python
#separating and computing average for each language

languages = {'en':'English','ja':'Japanese','de':'German','na':'Media','fr':'French','zh':'Chinese',
             'ru':'Russian','es':'Spanish'}


lang_sets = {}
avgs = {}

for lang in languages.keys():
    lang_sets[lang] = train_df[train_df.lang==lang].iloc[:,0:-1]
    avgs[lang] = lang_sets[lang].iloc[:,1:].sum(axis=0)/lang_sets[lang].shape[0]
```

```python
# plot for average no of views vs date for all languages


days = [r for r in range(avgs['en'].shape[0])]

fig = plt.figure(1,figsize=[15,10])
plt.ylabel('Views per Page')
plt.xlabel('Day')
plt.title('Pages in Different Languages')

for lang in languages.keys():
    plt.plot(days,avgs[lang],label = languages[lang])

plt.legend()
plt.show()
fig.savefig('views_comparison.png')
```

```python
# auto correlation and partial auto correlation

from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.stattools import acf

for key in avgs:
    fig = plt.figure(1,figsize=[10,5])
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    data = np.array(avgs[key])
    autocorr = acf(data)
    pac = pacf(data)

    x = [x for x in range(len(pac))]
    ax1.plot(x[1:],autocorr[1:])

    ax2.plot(x[1:],pac[1:])
    ax1.set_xlabel('Lag')
    ax1.set_ylabel('Autocorrelation')
    ax1.set_title(languages[key])

    ax2.set_xlabel('Lag')
    ax2.set_ylabel('Partial Autocorrelation')
    ax2.set_title(languages[key])
#     print(key)
    plt.show()
    fig.savefig(key+'_corr.png')
```

```python

# trend cycle seasonal
# hodrick prescott filter
import statsmodels.api as sm

hpfilter = sm.tsa.filters.hpfilter

for lang in ['es']:
    cycle,trend = hpfilter(avgs[lang],800)
    fig = plt.figure(1,figsize=[20,20])
    ax = [fig.add_subplot(311),fig.add_subplot(312),fig.add_subplot(313)]
    ax[0].plot(avgs[lang])
    ax[0].set_title(lang)
    ax[0].set_ylabel('Original Data')
    ax[0].set_xlabel('Day')
    ax[1].plot(trend)
    ax[1].set_title(lang)
    ax[1].set_ylabel('Trend')
    ax[1].set_xlabel('Day')
    ax[2].plot(cycle)
    ax[2].set_title(lang)
    ax[2].set_ylabel('Cycle')
    ax[2].set_xlabel('Day')
    plt.show()
```

```python
seasonal_decompose = sm.tsa.seasonal_decompose

for lang in ['en']:
    data = []
    for val in avgs[lang]:
        data.append(val)
    data = pd.DataFrame(data)
    dates = lang_sets[lang].columns
    dates = dates[1:]
    data.index = pd.to_datetime(list(dates))
    decomposition = seasonal_decompose(data)
    
    
    fig, ax = plt.subplots(ncols=1, nrows=4, sharex=True, figsize=(12,5))
    ax[0].set_ylabel('Observed')
    decomposition.observed.plot(ax=ax[0])
    ax[1].set_ylabel('Trend')
    decomposition.trend.plot(ax=ax[1])
    ax[2].set_ylabel('Seasonal')
    decomposition.seasonal.plot(ax=ax[2])
    ax[3].set_ylabel('Residual')
    decomposition.resid.plot(ax=ax[3])
    
    plt.tight_layout()
    plt.show()
    fig.savefig(lang+'_seasonal_decomp.png')
#     decomposition.plot()
```

```python
#fft
from scipy.fftpack import fft

days = [r for r in range(avgs['en'].shape[0])]

for lang in ['en','es']:
    fft_vals = fft(list(avgs[lang]))
    fft_abs = [np.sqrt(np.real(x)*np.real(x)+np.imag(x)*np.imag(x)) for x in fft_vals]
    xvals = [day/days[-1] for day in days]
    npts = len(xvals) // 2 + 1
    fft_abs = fft_abs[1:npts]
    xvals = xvals[1:npts]
    fig = plt.figure(1)
    plt.plot(xvals,fft_abs)
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.title('FFT - '+languages[lang])
    plt.show()
    fig.savefig(lang+'_fft.png')
```
